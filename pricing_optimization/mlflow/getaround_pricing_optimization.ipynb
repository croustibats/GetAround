```python
for col in df.select_dtypes('int'):
    plt.figure()
    sns.distplot(df[col])
```

```python
for col in df.select_dtypes('object'):
    print(f'{col :-<30} {df[col].unique()}')
```

```python
for col in df.select_dtypes('object'):
    plt.figure()
    df[col].value_counts().plot.pie()
```

```python
for col in df.select_dtypes('bool'):
    plt.figure()
    df[col].value_counts().plot.pie()
```

```python
for col in df.select_dtypes('bool'):
    df[col] = df[col].apply(lambda x: 1 if x else 0)
```
